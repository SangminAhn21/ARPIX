```python
text = "똥을 싸고 싶은데 휴지가 없네."
text = pg._preprocess(text)

beam = 5
temperature = 0.8
top_k = -1
top_p = 0.95
no_repeat_ngram_size = 4
len_penalty = 1.0

sampling = False

if top_k != -1 or top_p != -1:
    sampling = True

output = pg._model.translate(
    text,
    beam=beam,
    sampling=sampling,
    temperature=temperature,
    sampling_topk=top_k,
    sampling_topp=top_p,
    max_len_a=1,
    max_len_b=50,
    no_repeat_ngram_size=no_repeat_ngram_size,
    lenpen=len_penalty,
    num_return_sequences=3
    )
print(output)
output = pg._postprocess(output)
print(output)
```
